```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4), dpi=100)

ages_x=[10, 18, 25, 30, 35, 40, 45 ]
dev_y=[180, 270, 350, 400, 500, 550, 600 ]
plt.plot(ages_x, dev_y, 'ro-', linewidth=2, label='In USA')

py_dev_y=[200, 290, 390, 450, 540, 590, 660]
plt.plot(ages_x, py_dev_y, 'm--', linewidth=2, label='In UK')

uzb_dev_y=[40, 80, 200, 250, 350, 450, 500]
plt.plot(ages_x, uzb_dev_y, 'y>-', linewidth=2, label='In UZB')

plt.xlabel("Ages")
plt.ylabel("expenditure of money (USD) per month")
plt.title("expenditure of money (USD) per month by ages")

plt.legend()
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('fivethirtyeight')
x = [0, 1, 2, 3]
y = [0, 2, 4, 6]
plt.figure(figsize=(5,3), dpi=100)
plt.plot(x,y,'k--',linewidth=2, label='2x')

x2=np.arange(0,4.5, 0.5)
plt.plot(x2[:7], x2[:7]**2, 'yo-',linewidth=2, label='x^2')

plt.xticks([0,1,2,3,4,5])
plt.yticks([0,2,4,6,8,10])


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')

dev_x=[12, 16, 18, 25, 30, 35, 40]

x_index=np.arange(len(dev_x))
width=0.2

dev_y=[12, 14, 19, 23, 25, 29, 31]
plt.bar(x_index-width, dev_y,width=width,color='r',label='In UZB')

dex_y=[18, 20, 26, 22, 19, 15, 18]
plt.bar(x_index, dex_y,width=width,color='c',label='In USA')

dey_y=[15, 18, 22, 25, 26, 27, 33]
plt.bar(x_index+width, dey_y,width=width,color='m',label='In INDIA')

dez_y=[10, 13, 17, 21, 23, 25, 27 ]
plt.bar(x_index+width+width, dez_y,width=width,color='y',label='In Shweiz')

plt.xlabel('Ages')
plt.ylabel('percentage of crimes')
plt.title('percentage of crimes committed by ages in other countries')

plt.legend()
plt.xticks(ticks=x_index, labels=dev_x)
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4), dpi=100)

plt.style.use('fivethirtyeight')

slices=[82, 67, 83, 127, 144]
labels=['Uzbekistan', 'UK', 'Japan', 'Germany', 'USA']
explode=[0.1,0,0,0,0]

plt.pie(slices, labels=labels,explode=explode,shadow=True,autopct='%1.1f%%', wedgeprops={'edgecolor':'black'},)

plt.title('Car producing countries')
plt.tight_layout()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4), dpi=100)
plt.style.use('seaborn')

dev_x=[2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]

x_index=np.arange(len(dev_x))
width=0.2

dev_y=[500, 600, 650, 700, 900, 1000, 1100, 900]
plt.bar(x_index-width, dev_y,width=width,color='r',label='Germany')

dex_y=[300, 500, 600, 800, 900, 1200, 1300, 1000]
plt.bar(x_index, dex_y,width=width,color='c',label='UK')

dey_y=[600, 800, 900, 1200, 1500, 1300, 1500, 1200]
plt.bar(x_index+width, dey_y,width=width,color='m',label='USA')



plt.xlabel('years')
plt.ylabel('Sold cars (1=1000)')
plt.title('Cars selling ratings in tree countries by years')

plt.legend()
plt.xticks(ticks=x_index, labels=dev_x)
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
labels=['UZB','USA','UK']
values=[1.2, 5, 4.5]

bars=plt.bar(labels, values, color='c')
patterns=['x', '/', 'o']
for bar in bars:
    bar.set_hatch(patterns.pop())



plt.title('Incomes of countries (USD) per year in 2020')
plt.xlabel('Countries')
plt.ylabel('Income (USD) 1=1bln')

plt.show()
```
